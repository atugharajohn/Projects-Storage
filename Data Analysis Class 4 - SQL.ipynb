### CREATE CONNECTION

```python
#portnumber and username #supply the credential, and database you want to access
```

```python
engine = sqlalchemy.create_engine("mysql+pymysql://root:rootpassword@localhost:3306/quiz_data")
```

### READ A TABLE FROM THE DATABASE

#### Using SQL Table

```python
#Reading the people table
people = pd.read_sql_table("people", engine)
```

```python
people.head(5)
```

```python
people.nunique()
```

```python
people.shape
```

```python
#selecting all quiz points above 80

points_above_80 = people[people["quiz_points"] > 80]
points_above_80
```

### Using SQL Query

query = select * FROM states

```python
query = "SELECT * FROM states"
states = pd.read_sql_query(query, engine)
```

```python
#check the first five rows

states.head()
```

### Using real_sql 

```python
#using table name (people) with read_sql
table_name = pd.read_sql("people", engine)
table_name
```

```python
#using query with read_sql

query2 = "SELECT first_name, last_name, state_code FROM people"
using_query = pd.read_sql(query2, engine)
using_query
```

### SQL JOINS

```python
#join people table with the states table

query3 = "SELECT * FROM people ppl JOIN states st ON ppl.state_code = st.state_abbrev"

joined_data = pd.read_sql(query3, engine)
```

```python
joined_data
```

### JOIN PYTHON WAY

```python
#because the unique columns are different we use left on and right on
join_data = pd.merge(people, states, left_on = "state_code", right_on = "state_abbrev")
join_data.head()
```
