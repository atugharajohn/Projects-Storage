### Histogram

```python
#histogram - used to display distribution of numerical variables and not categorical
#groups your data into number of bins or groups
#and gives you count of data that falls into those bins
```

```python
#import the necessary libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
#inspect the distribution of the price. You can also change the default colour
sns.histplot(data = cars, x = "Price", bins=5, color = "orange");
plt.title("Price distribution of Cars")
plt.xlabel("Price of cars")
plt.ylabel("Frequency");
```

### Box plot - used to display five number summary
- minimum values
- maximum values
- lower quarter
- upper quarter
- median or middle number or 50th percent

Can be used for single variable, or bivariate analysis, a combination of categorical and numerical or just numerical, shows outliers or unusual values

```python
cars.head(2)
```

```python
#minimum values -
#maximum values
#lower quarter - 25% and above
#upper quarter - 75% and above
#median or middle number or 50th percent
```

```python
#boxplot of single variable
plt.figure(figsize = (10, 5))
sns.boxplot(data = cars, x = "Engine");
```

```python
#Bar chart is used to count the number of observations in each bin of a categorical variable
#we use countplot

sns.countplot(data = cars, x = "Transmission");
plt.title("Count of Transmission")
plt.xlabel("Transmission")
plt.ylabel("Frequency");
```

```python
sns.countplot(data = cars, x = "Fuel_Type");
```

```python
#rotating the x-labels
plt.figure(figsize = (8, 5))
g = sns.countplot(data = cars, x = "Fuel_Type")
g.set_xticklabels(g.get_xticklabels(), rotation=90, horizontalalignment='right')
plt.show()
```

```python
plt.figure(figsize = (7,5))
sns.countplot(data = cars, x = "Fuel_Type", hue = "Transmission");
```

## Bivariate Analysis
- This involves analyzing two variables to explore the relationship between them

```python
# we can make use of different plot- boxplot, categorical plot, scattered plot - single categorical and quantitative variable
```

```python
#number of seats vs engine
plt.figure(figsize = (12, 9))
sns.boxplot(data = cars, x = "Fuel_Type", y = "Engine");
```

### Scatter plot - For Two Numeric Variables

```python
#engine and price
plt.figure(figsize = (10, 6))
sns.scatterplot(data = cars, x = "Engine", y = "Price");
```

```python
#power and price (with hue as transmission)
plt.figure(figsize = (10, 6))
sns.scatterplot(data = cars, x = "Power", y = "Price", hue = "Transmission");
```

### Categorical Plot

Categorical and numerical

```python
#strip plot in catplot, so you can change the kind #bar, #violin, #strip plot
sns.catplot(data = cars, x = "Transmission", y = "Price", kind = "box")
```

```python
#using count plot
plt.figure(figsize=(10,6)) 
sns.countplot(x='Transmission', data=cars) 
plt.show()
```
